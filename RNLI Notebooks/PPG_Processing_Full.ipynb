```python
def plot_subfigure_raw_ppg(video_analytics: SessionAnalytics, axis: plt.Axes):
    axis.title.set_text("Raw PPG Data")
    axis.plot(video_analytics.stream_data_dictionary["PPG_GRN"]["Timestamp"], video_analytics.stream_data_dictionary["PPG_GRN"]["Value"])

def plot_subfigure_annotated_filtered_ppg(video_analytics: SessionAnalytics, axis: plt.Axes):
    axis.title.set_text("Filtered PPG Data")
    filtered_data = video_analytics.stream_data_dictionary["PPG_GRN_Filtered"].copy()
    systolic_peaks = video_analytics.stream_data_dictionary["PPG_GRN_Filtered_Systolic_Peaks"].copy()
    diastolic_trough = video_analytics.stream_data_dictionary["PPG_GRN_Filtered_Diastolic_Trough"].copy()
    diastolic_peaks = video_analytics.stream_data_dictionary["PPG_GRN_Filtered_Diastolic_Peaks"].copy()
    dicrotic_notch = video_analytics.stream_data_dictionary["PPG_GRN_Filtered_Dicrotic_Notch"].copy()

    axis.plot(filtered_data["Timestamp"], filtered_data["Value"])
    axis.scatter(systolic_peaks["Timestamp"], systolic_peaks["Value"], c="orange")
    axis.scatter(diastolic_trough["Timestamp"], diastolic_trough["Value"], c="blue")
    axis.scatter(diastolic_peaks["Timestamp"], diastolic_peaks["Value"], c="orange", marker="x")
    axis.scatter(dicrotic_notch["Timestamp"], dicrotic_notch["Value"], c="blue", marker="x")

def plot_subfigure_heart_rate(video_analytics: SessionAnalytics, axis: plt.Axes, previous_calibration_start_time: float, video_end_time: float, sample_duration: float = 20, sampling_step_count: float = 5):
    axis.title.set_text("Averaged Heart Rate")
    systolic_peaks = video_analytics.stream_data_dictionary["PPG_GRN_Filtered_Systolic_Peaks"]
    bpm_dataframe = pd.DataFrame(columns=["Timestamp", "BPM"])

    #region [ Calculate Average Heartrate ]
    min_bpm: float = float("inf")
    max_bpm: float = 0

    sample_time = previous_calibration_start_time + (sample_duration / 2)
    while sample_time < (video_end_time - (sample_duration / 2)):
        peaks_subset = systolic_peaks[
            (systolic_peaks["Timestamp"] > (sample_time - (sample_duration / 2))) &
            (systolic_peaks["Timestamp"] < (sample_time + (sample_duration / 2)))
        ]

        bpm = len(peaks_subset) * (60 / sample_duration)

        if bpm < min_bpm: min_bpm = bpm
        if bpm > max_bpm: max_bpm = bpm

        bpm_dataframe.loc[len(bpm_dataframe)] = [
            sample_time,
            bpm
        ]

        print(f"Sample Time: {sample_time}, BPM: {bpm}")

        sample_time = sample_time + sampling_step_count
    #endregion

    #region [ Calculate Baseline ]
    baseline_sample_subset = bpm_dataframe[
        (bpm_dataframe["Timestamp"] > -40) &
        (bpm_dataframe["Timestamp"] < 0)
    ]

    mean_baseline = baseline_sample_subset["BPM"].mean()
    #endregion

    #region [ Plot Data ]
    axis.set_ylim(40, 120)
    duration = video_end_time - previous_calibration_start_time

    # Plot Baseline Heart Rate
    axis.hlines(mean_baseline, previous_calibration_start_time, video_end_time, colors="red", linestyles="dashed", alpha=0.25)

    # Plot Min and Max
    axis.hlines(min_bpm, previous_calibration_start_time, video_end_time, colors="gray", linestyles="dashed", alpha=0.25)
    axis.hlines(max_bpm, previous_calibration_start_time, video_end_time, colors="gray", linestyles="dashed", alpha=0.25)

    #Plot Key Time Points
    axis.axvspan(xmin= -40, xmax= 0, color="red", alpha=0.05)

    # Plot BPM Data
    axis.plot(bpm_dataframe["Timestamp"], bpm_dataframe["BPM"])
    #endregion

def plot_subfigure_raw_gyro(video_analytics: SessionAnalytics, axis: plt.Axes):
    axis.title.set_text("Gyroscope Data")
    axis.plot(video_analytics.stream_data_dictionary["GYRO_X"]["Timestamp"], video_analytics.stream_data_dictionary["GYRO_X"]["Value"])
    axis.plot(video_analytics.stream_data_dictionary["GYRO_Y"]["Timestamp"], video_analytics.stream_data_dictionary["GYRO_Y"]["Value"])
    axis.plot(video_analytics.stream_data_dictionary["GYRO_Z"]["Timestamp"], video_analytics.stream_data_dictionary["GYRO_Z"]["Value"])

def plot_subfigure_raw_accelerometer(video_analytics: SessionAnalytics, axis: plt.Axes):
    axis.title.set_text("Accelerometer Data")
    axis.plot(video_analytics.stream_data_dictionary["ACC_X"]["Timestamp"], video_analytics.stream_data_dictionary["ACC_X"]["Value"])
    axis.plot(video_analytics.stream_data_dictionary["ACC_Y"]["Timestamp"], video_analytics.stream_data_dictionary["ACC_Y"]["Value"])
    axis.plot(video_analytics.stream_data_dictionary["ACC_Z"]["Timestamp"], video_analytics.stream_data_dictionary["ACC_Z"]["Value"])

def generate_session_figures(session_analytics: SessionAnalytics):
    marker_pairs = session_analytics.get_paired_markers("Marker", "VideoStart", "VideoEnd")

    for video_index, row in marker_pairs.iterrows():
        #region [ Get Data ]
        video_start_time = row["Start Timestamp"]
        video_end_time = row["End Timestamp"]
        previous_calibration_start_time = video_start_time - 60
        next_calibration_end_time = video_end_time + 60

        video_analytics = session_analytics.get_filtered_subset([TimestampStreamFilter(previous_calibration_start_time, next_calibration_end_time)])
        video_analytics.localise_timestamps(video_start_time)

        video_end_time = video_end_time - video_start_time
        previous_calibration_start_time = previous_calibration_start_time - video_start_time
        next_calibration_end_time = next_calibration_end_time - video_start_time
        video_start_time = 0

        figure_duration = video_end_time - previous_calibration_start_time
        #endregion

        figure, axes = plt.subplots(nrows=5, ncols=1, figsize=(figure_duration * 0.5, 20), dpi=300)
        for _, axis in enumerate(axes): axis.set_xlim(previous_calibration_start_time, video_end_time)

        plot_subfigure_raw_ppg(video_analytics, axes[0])
        plot_subfigure_annotated_filtered_ppg(video_analytics, axes[1])
        plot_subfigure_heart_rate(video_analytics, axes[2], previous_calibration_start_time, next_calibration_end_time)
        plot_subfigure_raw_gyro(video_analytics, axes[3])
        plot_subfigure_raw_accelerometer(video_analytics, axes[4])

        #Mark Video Start and End
        print(f"Adding Video Start Time Marker: {video_start_time}")
        for _, axis in enumerate(axes): axis.vlines(video_start_time, axis.get_ylim()[0], axis.get_ylim()[1], colors="black", alpha=0.5)

        print(f"Adding Video End Time Marker: {video_end_time}")
        for _, axis in enumerate(axes): axis.vlines(video_end_time, axis.get_ylim()[0], axis.get_ylim()[1], colors="black", alpha=0.5)

        figure.suptitle(f"Participant {session_analytics.file_name}: Video {video_index + 1} Full PPG Data.", fontsize=32)
        figure.tight_layout(rect=[0.005, 0.01, 0.995, 0.975])

        os.makedirs(f"V:/Exports/Participant {session_analytics.file_name}", exist_ok=True)
        plt.savefig(f"V:/Exports/Participant {session_analytics.file_name}/Video {video_index + 1} Full PPG Data.png")
        plt.close()
```

```python
def process_all_sessions():
    for _, row in _experiment_analytics.analytics_dataframe.iterrows():
        session_analytics: SessionAnalytics = row["Analysis_Object"]
        session_analytics.localise_timestamps()

        StreamProcesses.butterworth_filter(session_analytics, "PPG_GRN", "PPG_GRN_Filtered")
        StreamProcesses.detect_ppg_peaks(session_analytics, "PPG_GRN_Filtered")

        generate_session_figures(session_analytics)

process_all_sessions()
```
